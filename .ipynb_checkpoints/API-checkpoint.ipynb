```python
import snscrape.modules.twitter as sntwitter
```

```python
def get_tweets(query, max_tweets):
    tweets = []
    for i, tweet in enumerate(sntwitter.TwitterSearchScraper(query).get_items()):
        tweets.append(tweet)
        
        # Break the loop if the desired number of tweets is reached
        if i + 1 == max_tweets:
            break
    return tweets
```

```python
def weather(key = '7645c564ea36454c8e4e03c788702956'):
    import requests
    api_url =  f'https://api.weatherbit.io/v2.0/current?lat=35.7796&lon=-78.6382&key={key}&include=minutely'
    response = requests.get(api_url)
    data = response.json()
    
    precip = []
    snow = []
    temp = []
    ts = []
    time_local = []
    time_utc = []

    for i in range(0,len(data['minutely'])):
            z = data['minutely'][i]
            precip.append(z['precip'])
            snow.append(z['snow'])
            temp.append(z['temp'])
            ts.append(z['ts'])
            time_local.append(z['timestamp_local'])
            time_utc.append(z['timestamp_utc'])

    import pandas as pd

    df = pd.DataFrame({'Rain':precip,'Snow': snow,'Temperatur': temp,
                       'Time Stamp': ts,'Local Time': time_local,'Time UTC': time_utc})
    return df
```

```python
weather()
```

```python
df.to_csv('Weather_dataframe.csv', index=False)
```
