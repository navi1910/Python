```python
x = np.linspace(0, 5, 10)
y = np.arange(0, 10)**2
```

```python
plt.plot(x, y, color = 'g')
plt.title('This is my plot')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.show()
```

```python
plt.subplot(1,2,2)
plt.plot(x, y, color = 'r')
plt.title('X vs Y')

plt.subplot(1,2,1)
plt.plot(y, x, color = 'blue')
plt.title('Y vs X')
plt.show()
```

```python
plt.subplot(2,1,2)
plt.plot(x, y, color = 'r')
plt.title('X vs Y')

plt.subplot(2,1,1)
plt.plot(y, x, color = 'blue')
plt.title('Y vs X')
plt.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #l, b, w, h
```

```python
# Object Oriented Plotting in Python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, color = 'r')
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.4, 0.3, 0.4])

axes1.plot(x, y, color = 'r')
axes2.plot(y, x, color = 'g')
plt.show()
```

```python
fig2 = plt.figure()
axes = fig2.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_title('X vs Y')
axes.set_xlabel('X')
axes.set_ylabel('Y')

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (10, 6))

axes[0].plot(x, y)
axes[1].plot(y, x, color = 'r')
axes[0].set_title('plot')

plt.show()
```

```python
fig = plt.figure(figsize = (10, 6))

a1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

a1.plot(x, y, marker = 'o', lw = 2.5, c = 'r', mec = 'g', mfc = 'y', label = '1st', alpha = 0.1)
a1.plot(x,x , marker = 'x', mec = 'g', label = '2nd', ls = ':')

a1.legend(loc = 2)
plt.tight_layout()
plt.show()
```
